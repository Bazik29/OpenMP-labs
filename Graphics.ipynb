```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

raw_data = pandas.read_csv('test.csv')

data = raw_data.groupby(['Threads', 'Dimension'], as_index = False).mean()
#data = data[data.Dimension < 2000]

plt.title('Время выполнения')
for i in sorted(set(data['Dimension'])):
    x = data.loc[data.Dimension == i]
    plt.plot(x['Threads'], x['RunTime'], marker=".", label="Матрица: {dim}x{dim}".format(dim = i))
plt.legend()
plt.xlabel('Число потоков')
plt.ylabel('Среднее время выполнения сек')
plt.show()

plt.title('Ускорение')
for i in sorted(set(data['Dimension'])):
    x = data.loc[data.Dimension == i]
    one_tread_time = float(x[x.Threads == 1]['RunTime'])
    plt.plot(x['Threads'], one_tread_time / x['RunTime'], marker=".", label="Матрица: {dim}x{dim}".format(dim = i))
plt.legend()
plt.xlabel('Число потоков')
plt.ylabel('Ускорение')
plt.show()

plt.title('Эффективность')
for i in sorted(set(data['Dimension'])):
    x = data.loc[data.Dimension == i]
    one_tread_time = float(x[x.Threads == 1]['RunTime'])
    plt.plot(x['Threads'], one_tread_time / x['RunTime'] / x['Threads'], marker=".", label="Матрица: {dim}x{dim}".format(dim = i))
plt.legend()
plt.xlabel('Число потоков')
plt.ylabel('Эффективность')
plt.show()
```
